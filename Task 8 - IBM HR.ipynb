#### Q10: What is the relationship between the over time and the attrition?

```python
ibm1=ibm.iloc[:,[1,22]]
ibm1['OverTime'].replace('No',0,inplace=True)
ibm1['OverTime'].replace('Yes',1,inplace=True)
sns.displot(data = ibm1, x = 'OverTime', hue = 'Attrition', kind="kde",palette='rocket')
```

#### Ans: Zero represents that there is no over time and One represents that there is over time. The higher the over time the higher the probability for the employee to get attrited.

####   
#### Q11: How does the age contributes in the attrition?

```python
sns.displot(data = ibm, x = 'Age', hue = 'Attrition', kind="kde",palette='rocket')
```

```python
ibm2=ibm.iloc[:,:2]
ibm2['Attrition'].replace('Yes',1,inplace=True)
ibm2['Attrition'].replace('No',0,inplace=True)
sns.set(rc = {'figure.figsize':(8,4)})
cmap = sns.cubehelix_palette(as_cmap=True)
ax = sns.heatmap(ibm2.corr(),annot = True, cmap=cmap)
```

#### Ans: From the heatmap and the distribution, we can conclude that there is no relationship between the age and the attrition.
